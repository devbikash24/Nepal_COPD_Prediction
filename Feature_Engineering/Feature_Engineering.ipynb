## **Before starting with Feature Engineering. Let's Understand the data.**

```python
df=pd.read_csv("../data/synthetic_COPD_data.csv")
```

```python
df.head()
```

```python
print(df['Age'].nunique()) # Let be numerical itself.But we can convert it to categorical with binning.
```

```python
bins = [0, 30, 45, 60, 75, 100]
labels=['young','adult','middle_aged','old','too_old']
df['Age_Category'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
df.head()
```

### We can do similar work for BMI categories and Air_Poulltion_level categories.

![A beautiful landscape](../data/bmi__WHO.jpg)

```python
bins=[0,19,25,30,35,40]
labels=['underweight','normal','overweight','obese','too_obese']
df['BMI_category']=pd.cut(df['BMI'],bins=bins,labels=labels,right=False)
```

```python
df.head()
```


![A beautiful landscape](../data/air_pollution_standards_cpcb.png)

```python
bins=[0,50,100,200,300,400,500]
labels=['Good','Satisfactory','Moderate','Poor','Very_Poor','Severe']
df['Air_Pollution_Level_category']=pd.cut(df['Air_Pollution_Level'],bins=bins,labels=labels,right=False)
```

```python
df.head()
```

# now for Smoking_status we can use Ordinal_Encodin when we are training the data.
