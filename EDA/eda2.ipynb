## Key Contributors to COPD in Nepal

   Indoor Air Pollution: The use of biomass fuels (wood, dung, crop residues) for cooking and heating releases harmful pollutants, which can irritate and damage the lungs over time, significantly increasing the risk of COPD.

   Outdoor Air Pollution: High levels of air pollution in urban areas, particularly from vehicle emissions and industrial activities, contribute to respiratory issues and exacerbate COPD symptoms.

   Tobacco Use: Smoking and the use of other forms of tobacco, such as khaini, are major risk factors for COPD, directly damaging the lungs and impairing respiratory function.

   Socioeconomic Factors: Limited access to healthcare, poor living conditions, and higher exposure to indoor pollutants are more common among lower socioeconomic groups, heightening their risk of COPD.

   Occupational Exposure: Jobs in agriculture, construction, and mining expose workers to dust, chemicals, and other lung irritants, increasing the likelihood of developing COPD.

   Genetic Factors: Genetic predispositions, such as alpha-1 antitrypsin deficiency, can increase the susceptibility to COPD, although these factors are less studied in Nepal.

   Infections: Frequent childhood respiratory infections, common in areas with limited healthcare, can damage the lungs and increase COPD risk later in life.

   Climate and Geography: High-altitude living and diverse geographic conditions can pose unique respiratory challenges, though COPD is primarily influenced by other environmental and lifestyle factors.

Addressing these contributors requires targeted interventions, improved healthcare access, and policies to reduce pollution and promote healthier living conditions.

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting styles for the plots
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
# Loading the data
df = pd.read_csv("../data/synthetic_COPD_data.csv")
```

```python
df.head()
```

```python
# Overview of the data
print("Data Overview:")
print(df.info())
```

```python
print("\nSummary Statistics:")
df.describe()
```

```python
# Checking for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Univariate Analysis: Distribution of Numerical Features
numerical_features = df.select_dtypes(include=[np.number]).columns
for col in numerical_features:
    plt.figure()
    sns.histplot(df[col], kde=True, bins=30, color='blue')
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Univariate Analysis: Distribution of Categorical Features
categorical_features = df.select_dtypes(include=['object']).columns
for col in categorical_features:
    plt.figure()
    sns.countplot(x=col, data=df, palette="Set2", hue="COPD_Diagnosis")
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Bivariate Analysis: Scatter plots and Pair plots
sns.pairplot(df, hue='COPD_Diagnosis')  
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```

```python
# Boxplot for Numerical Features across Categorical Variables
for col in numerical_features:
    for cat_col in categorical_features:
        plt.figure()
        sns.boxplot(x=cat_col, y=col, data=df, palette="coolwarm", hue="COPD_Diagnosis")
        plt.title(f"Boxplot of {col} vs {cat_col}")
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Selecting only numerical columns
numerical_df = df.select_dtypes(include=[np.number])

# Correlation Analysis for numerical data
plt.figure(figsize=(12, 10))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap for Numerical Data")
plt.show()
```

```python
# Analyzing Relationships with the Target Variable
target_col = 'COPD_Diagnosis'  # Replace with the actual target column name
for col in numerical_features:
    plt.figure()
    sns.boxplot(x=target_col, y=col, data=df, palette="viridis")
    plt.title(f"{col} Distribution by {target_col}")
    plt.show()
```
